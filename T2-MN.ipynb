Ejercicio planteado

```python
from scipy.optimize import newton

def f(x):
    y = x**3 - 3 * x**2 + x - 1
    print(f"f({x}) = {y}")
    return y
def f_prime(x):
    return 3 * x**2 - 6 * x + 1
estimations = []
def callback(x):
    estimations.append(x)

def newton_with_estimations(func, x0, fprime, tol=1e-6, maxiter=100):
    x = x0
    for _ in range(maxiter):
        callback(x)  
        fx = func(x)
        fpx = fprime(x)
        
        if abs(fpx) < 1e-10:
            raise RuntimeError("La derivadacero")
        
        x_new = x - fx / fpx
        if abs(x_new - x) < tol:
            callback(x_new)  
            return x_new
        x = x_new
    
    raise RuntimeError("No se encontr√≥ la ra√≠z despu√©s de las iteraciones m√°ximas")

# Estimaci√≥n inicial
x0 = 100
root = newton_with_estimations(f, x0, f_prime)

print("Ra√≠z encontrada:", root)
print("Estimaciones visitadas:", estimations)
```
